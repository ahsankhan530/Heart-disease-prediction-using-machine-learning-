```python
Y=df.target.values
x_data=df.drop(['target'],axis=1)
```

```python
#normalize
x=(x_data-np.min(x_data))/(np.max(x_data)-np.min(x_data)).values
```

```python
#We will split our data. 80% of our data will be train data and 20% of it will be test data.
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,Y,test_size=0.2,random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
model=nb.fit(x_train, y_train)
print("Accuracy of Naive Bayes: {:.2f}%".format(nb.score(x_test,y_test)*100))
```

```python
print(model.predict([[63,1,3,145,233,1,0,150,0,2.3,0,0,1]]))
```

```python
# Random Forest Classification
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 1000, random_state = 1)
modelf=rf.fit(x_train, y_train)
print("Random Forest Algorithm Accuracy Score : {:.2f}%".format(rf.score(x_test,y_test)*100))
```

```python
import matplotlib.pyplot as plt
import random

inside = 0
n = 10**3

x_inside = []
y_inside = []
x_outside = []
y_outside = []

for _ in range(n):
    x = random.uniform(-1.0,1.0)
    y = random.uniform(-1.0,1.0)
    if x**2+y**2 <= 1:
        inside += 1
        x_inside.append(x)
        y_inside.append(y)
    else:
        x_outside.append(x)
        y_outside.append(y)
```

```python
pi = 4*inside/n
print(pi)
```

```python
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.scatter(x_inside, y_inside, color='g', marker='s')
ax.scatter(x_outside, y_outside, color='r', marker='s')
fig.show()
```
